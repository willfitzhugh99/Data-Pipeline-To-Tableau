```python
#scrape all games in in a year for all years 2002 to 2022:
#export csv file for each year so process can be done in parts

#for each year in given range (range is inclusive), a list of dates is created using datetime module, then
#scrape_ESPN function scrapes data for each date in year and data is concatenated then exported

from datetime import date, timedelta

def scrape_years(year_start, year_end):

    for year in range(year_start,year_end + 1):
        start_date = date(year, 7, 2)  #season starts and ends in July
        end_date = date(year+1, 7, 1)

        delta = end_date - start_date   # returns timedelta

        scrape_dates = []
        for i in range(delta.days + 1):
            day = start_date + timedelta(days=i)
            scrape_dates.append(str(day).replace('-',''))
    
    
        data = scrape_ESPN(str(year)+'0701')   # create df from date 07
        for i in scrape_dates:
            holder = scrape_ESPN(i)
            data = pd.concat([data,holder])
    
        label = 'Data/data' + str(year) + '.csv'
        data.to_csv(label, index=False)
```

```python
scrape_years(2002, 2003)
```

```python
# aggegation example

data03 = pd.read_csv('Data/data2003.csv')
spa03 = data03[ data03.League == 'Italian Serie A']

home_results = spa03.groupby('Home Team').sum()[['Home Points']].reset_index()
away_results = spa03.groupby('Away Team').sum()[['Away Points']].reset_index()

spa03_table = pd.DataFrame(
                pd.merge(home_results, away_results, how = 'left', left_on = 'Home Team', right_on = 'Away Team'
                      ).set_index('Home Team'
                                 ).drop(columns = ['Away Team']
                                       ).sum(axis=1)
            ).rename(columns = {0:'Points'}
                    ).sort_values(by = 'Points', ascending = False)
    
spa03_table
```
