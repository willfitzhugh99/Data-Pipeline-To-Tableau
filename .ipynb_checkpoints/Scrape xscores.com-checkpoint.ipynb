```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

##### xscores.com url format: 
https://www.xscores.com/soccer/<font color='blue'>*country*</font>/<font color='blue'>*league_name*</font>/results/<font color='blue'>*start_year*</font>-<font color='blue'>*end_year*</font>

```python
league_names = pd.DataFrame({
    'country':['spain', 'england','germany','france','italy','europe-uefa','europe-uefa','spain','england',
               'germany','france', 'italy'],
    'league_name':['primera-division','premier-league','bundesliga','ligue-1','serie-a','uefa-champions-league',
    'uefa-europa-league','fa-cup','fa-cup','fa-cup','fa-cup','fa-cup']
})
```

```python
league_names.iloc[0,:][1]
```

```python
country = league_names.iloc[0,:][0]
league_name = league_names.iloc[0,:][1]
start_year = '2021'
end_year = '2022'
url = 'https://www.xscores.com/soccer/' + country + '/' + league_name + '/results/' + start_year + '-' + end_year
html_content = requests.get(url).text
soup = BeautifulSoup(html_content, "html.parser")
soup
```

```python
tags = ['a']
classes = [
    'match wrapper'
]
scraper = soup.select('div')


stuff = []
for i in scraper:
    stuff.append(i.get_text())
stuff
```
