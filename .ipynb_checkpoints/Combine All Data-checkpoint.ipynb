```python
import pandas as pd
```

```python
league_names = pd.DataFrame({
    'country':['spain', 'england','germany','france','italy'],
    'league_name':['primera-division','premier-league','bundesliga','ligue-1','serie-a']
})
cup_names = pd.DataFrame({
    'country':['europe-uefa','europe-uefa','spain','england','germany','france', 'italy'],
    'league_name':['uefa-champions-league','uefa-europa-league','fa-cup','fa-cup','fa-cup','fa-cup','fa-cup']
})
leagues_cups = pd.concat([league_names,cup_names])
leagues_cups
```

```python
i = 0
pd.read_csv('Combined Data/' + leagues_cups.iloc[i,0] + '_' + leagues_cups.iloc[i,1] + '_all_games.csv')
```

```python
#combine upcoming fixtures with completed games
#then duplicate data and rename columns
#effectively changing an observation from a single game to each teams involvement in a game (1 row per game vs 2)

for i in range(len(leagues_cups)):
    label1 = 'Combined Data/' + leagues_cups.iloc[i,0] + '_' + leagues_cups.iloc[i,1] + '_all_games.csv'
    label2 = 'Upcoming Fixtures/' + leagues_cups.iloc[i,0] + '_' + leagues_cups.iloc[i,1] + '_upcoming_fixtures.csv'
    
    completed = pd.read_csv(label1)
    upcoming = pd.read_csv(label2)
    
    data = pd.concat([completed,upcoming])
    
    home = data.copy()
    away = data.copy()
    
    home = data.rename(columns = {
        'Home_Team':'Team',
        'Away_Team':'Opponent',
        'Home_Score':'Team_Score',
        'Away_Score':'Opponent_Score',
        'Home_Points':'Team_Points',
        'Away_Points':'Opponent_Points'
    })
    home['Location'] = 'Home'
    
    away = away.rename(columns = {
        'Away_Team':'Team',
        'Home_Team':'Opponent',
        'Away_Score':'Team_Score',
        'Home_Score':'Opponent_Score',
        'Away_Points':'Team_Points',
        'Home_Points':'Opponent_Points'
    })
    away['Location'] = 'Away'
    
    label3 = 'Tab Data/' + leagues_cups.iloc[i,0] + '_' + leagues_cups.iloc[i,1] + '_duplicated_games.csv'
    pd.concat([home,away]).to_csv(label3,index=False)
```

```python
i=1
pd.read_csv('Tab Data/' + leagues_cups.iloc[i,0] + '_' + leagues_cups.iloc[i,1] + '_duplicated_games.csv')
```

```python
#combine all data onto single file, with league indicator
data = pd.DataFrame(columns=['Round', 'Date', 'Time', 'Team', 'Team_Score', 'Opponent_Score', 'Opponent', 
                    'Home_Score_AET', 'Away_Score_AET', 'Home_Penalties','Away_Penalties', 'Team_Points',
                    'Opponent_Points', 'season', 'Location'])

for i in range(len(leagues_cups)):        
    label = 'Tab Data/' + leagues_cups.iloc[i,0] + '_' + leagues_cups.iloc[i,1] + '_duplicated_games.csv'
    df_to_add = pd.read_csv(label)
    df_to_add['Country'] = leagues_cups.iloc[i,0]
    df_to_add['Competition'] = leagues_cups.iloc[i,1]
    
    data = pd.concat([data,df_to_add])
        
```

```python
data.to_csv('Full_Dataset.csv',index=False)
```
